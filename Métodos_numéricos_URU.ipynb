Información de la asignatura $$\int f(x) dx$$

```python
# @title
import numpy as np
import matplotlib.pyplot as plt
import math
from ipywidgets import interact, IntSlider, FloatSlider
import ipywidgets as widgets

def plot_taylor_approximation(x_val, n_terms):
    """
    x_val: El valor de x para evaluar e^x
    n_terms: Número de términos de la serie de Taylor a utilizar
    """

    # 1. Valor Verdadero
    true_val = math.exp(x_val)

    # 2. Aproximación (Suma de términos)
    approx_val = 0
    terms = []

    # Calculamos la serie término a término
    for i in range(n_terms):
        term = (x_val**i) / math.factorial(i)
        approx_val += term
        terms.append(approx_val)

    # 3. Cálculo de Errores
    abs_error = true_val - approx_val
    rel_error_percent = abs((true_val - approx_val) / true_val) * 100

    # --- VISUALIZACIÓN ---
    plt.figure(figsize=(12, 4))

    # Gráfica 1: Convergencia visual
    plt.subplot(1, 2, 1)
    plt.plot(range(1, n_terms + 1), terms, 'o-', label='Aproximación', color='orange')
    plt.axhline(y=true_val, color='blue', linestyle='--', label=f'Valor Verdadero ($e^{{{x_val}}}$)')
    plt.title(f'Convergencia de la Serie de Maclaurin para $e^{{{x_val}}}$')
    plt.xlabel('Número de Términos')
    plt.ylabel('Valor Calculado')
    plt.grid(True, alpha=0.3)
    plt.legend()

    # Gráfica 2: Reducción del Error
    plt.subplot(1, 2, 2)
    # Calculamos el error para cada paso para graficar la caída
    errors = [abs((true_val - t)/true_val)*100 for t in terms]
    plt.plot(range(1, n_terms + 1), errors, 'r-o')
    plt.title('Error Relativo Porcentual (%)')
    plt.xlabel('Número de Términos')
    plt.ylabel('% Error (Escala Log)')
    plt.yscale('log') # Escala logarítmica para ver mejor errores pequeños
    plt.grid(True, which="both", ls="-", alpha=0.3)

    plt.tight_layout()
    plt.show()

    # --- DASHBOARD DE RESULTADOS ---
    print(f"📊 REPORTE DE RESULTADOS:")
    print("-" * 30)
    print(f"Valor Verdadero (e^{x_val}): {true_val:.10f}")
    print(f"Valor Aproximado ({n_terms} términos): {approx_val:.10f}")
    print("-" * 30)
    print(f"❌ Error Absoluto: {abs_error:.10e}")
    print(f"📉 Error Relativo: {rel_error_percent:.10f}%")

# Crear el Dashboard Interactivo
print("Mueve los deslizadores para ver cómo el número de términos afecta el error:")
interact(plot_taylor_approximation,
         x_val=FloatSlider(value=1.5, min=0.1, max=5.0, step=0.1, description='Valor de x:'),
         n_terms=IntSlider(value=1, min=1, max=15, step=1, description='Nº Términos:'));
```
